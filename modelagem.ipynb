```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import linearmodels.iv as iv
from linearmodels.panel import PanelOLS
from linearmodels.panel import RandomEffects
```

```python
df = pd.read_parquet("dados_agrupados/censo_agrupado_por_regioes_imediatas.gzip")
```

```python
df.describe()
```

```python
df.query("id_regiao_geografica_imediata == 110001").T
```

```python
results = smf.ols("log_depositos_totais_deflacionada_per_capita ~ razao_dependencia_idosos + branco", data=df).fit()
print(results.summary())
```

```python
results = smf.ols("log_depositos_totais_deflacionada_per_capita ~ razao_dependencia_idosos + log_renda_dom_deflacionada_per_capita +sit_setor_C+filhos_nasc_vivos+ branco + anos_estudoC+sanitario+ nasceu_mun + anos_mor_mun + tx_casado + ocup_propria + n_pes_dom", data=df).fit()
print(results.summary())
```

```python
results = PanelOLS.from_formula("log_depositos_totais_deflacionada_per_capita ~ 1 + razao_dependencia_idosos + log_renda_dom_deflacionada_per_capita +sit_setor_C+filhos_nasc_vivos+ branco + anos_estudoC+sanitario+ nasceu_mun + anos_mor_mun + tx_casado + ocup_propria + n_pes_dom + EntityEffects + TimeEffects", data=df.set_index(["dt_year", "id_regiao_geografica_imediata"])).fit()
print(results)
```

```python
results = RandomEffects.from_formula("log_depositos_totais_deflacionada_per_capita ~ razao_dependencia_idosos + log_renda_dom_deflacionada_per_capita +sit_setor_C+filhos_nasc_vivos+ branco + anos_estudoC+sanitario+ nasceu_mun + anos_mor_mun + tx_casado + ocup_propria + n_pes_dom", data=df.set_index(["dt_year", "id_regiao_geografica_imediata"])).fit()
print(results)
```
